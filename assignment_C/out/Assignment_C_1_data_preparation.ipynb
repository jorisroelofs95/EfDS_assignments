```python
print(f"Adjustments I will make to the PL dataset, based on the information:")
print(f"-Height and weight have one decimal. I will changed it from floating to integer to be consistent with NL dataset.")
print(f"-There are empty values in column sex. I want them removed, also from all other columns that contain NaN values. Apart from smokes column. There i will change the value to unknown")
print(f"-Smokes column should be changed to yes/no instead of true/false to be consistent with NL dataset")
print(f"-The column name heigth is wrongly called heigth. I will change it to height")
```

```python
PL_changed = PL.copy() # I made a copy of the dataframe. I will change/update this one below

PL_changed["smokes"] = PL_changed["smokes"].fillna("Unknown")
PL_changed = PL_changed.dropna(axis=0)
PL_changed = PL_changed.rename(columns={"heigth": "height"})
PL_changed.loc[PL_changed["smokes"] == False, "smokes"] = "no"
PL_changed.loc[PL_changed["smokes"] == True, "smokes"] = "yes"
PL_changed["height"] = PL_changed["height"].astype(int)
PL_changed["weight"] = PL_changed["weight"].astype(int)
PL_changed["country"] = "PL"

PL_changed
```

```python
print(UK)
UK.keys()
print(f"the maximum weight is: {max(UK["weight"])}") # now i want to check if all the data in height and weight is logical. There shouldnt be data above 300 cm or weight above 250 kg
print(f"the minimum weight is: {min(UK["weight"])}")
print(f"the maximum height is: {max(UK["height"])}")
print(f"the minimum height is: {min(UK["height"])}")
print(UK["sex"].unique())
print(UK["smokes"].unique())
print(UK.info())
print(UK.describe())
```

```python
print(f"I will change the following things in the UK dataset to harmonize and clean the data:")
print(f"-change the 0 and 1 values in the smokes column to yes and no")
print(f"-harmonize all the values in the sex column to only female or male")
print(f"-the column name height should be weight and vice versa, considering the data values. I will change this")
print(f"-I will make the values of height and weight integers, just like in the NL data")
print(f"-Any values that don't make sense in height and weight, I will exclude. So if weight is lower than 25 kg or higher than 650. This is simply not possible.")
print(f"-Also height higher than 300 and lower than 50 (like -1) i will discard")
UK_changed = UK.copy()

UK_changed = UK_changed.dropna(axis=0)
UK_changed["smokes"] = UK_changed["smokes"].astype(str)
UK_changed.loc[UK_changed["smokes"] == "0", "smokes"] = "no" # assuming 0 is equal to "no"
UK_changed.loc[UK_changed["smokes"] == "1", "smokes"] = "yes"

UK_changed.loc[UK_changed["sex"] == "FEMALE", "sex"] = "female"
UK_changed.loc[UK_changed["sex"] == "Woman", "sex"] = "female"
UK_changed.loc[UK_changed["sex"] == "Male", "sex"] = "male"
UK_changed.loc[UK_changed["sex"] == "MALE", "sex"] = "male"
UK_changed.loc[UK_changed["sex"] == "Female", "sex"] = "female"

UK_changed = UK_changed.rename(columns={"height": "weight1"})
UK_changed = UK_changed.rename(columns={"weight": "height"})
UK_changed = UK_changed.rename(columns={"weight1": "weight"})

UK_changed["height"] = UK_changed["height"].astype(int)
UK_changed["weight"] = UK_changed["weight"].astype(int)

UK_changed = UK_changed[(UK_changed["weight"] < 650) & (UK_changed["weight"] > 25) & (UK_changed["height"] > 50) & (UK_changed["height"] < 300)]

UK_changed["country"] = "UK"
UK_changed
```

```python
print(US)
US.keys()
print(f"the maximum weight is: {max(US["weight"])}")
print(f"the minimum weight is: {min(US["weight"])}")
print(f"the maximum height is: {max(US["height"])}")
print(f"the minimum height is: {min(US["height"])}")
print(US["sex"].unique())
print(US["smokes"].unique())
print(US.info())
print(US.describe())
```

```python
print(f"I will change the following things in the US dataset to harmonize and clean the data:")
print(f"-I will drop all the NaN values, given they are not in the columngs sex and smokes")
print(f"-The column height becomes integer values")

US_changed = US.copy()

US_changed = US_changed.dropna(axis=0)
US_changed["country"] = "US"
US_changed["height"] = US_changed["height"].astype(int)

print(f"\nThe maximum weight after dropping nan values is: {max(US_changed["weight"])}")
print(f"The minimum weight after dropping nan values is: {min(US_changed["weight"])}")
print(f"The maximum height after dropping nan values is: {max(US_changed["height"])}")
print(f"The minimum height after dropping nan values is: {min(US_changed["height"])}")

US_changed
```

```python
print(f"I will combine the four changed datasets and add the bmi formula to the data: bmi = weight (kg) / (height in meter * height in meter)")

countries_comb = pd.concat([NL_changed, PL_changed, UK_changed, US_changed])

countries_comb["bmi"] = countries_comb["weight"] / ( (countries_comb["height"]/100)*(countries_comb["height"]/100) )

countries_comb = countries_comb.round({"bmi": 2})
countries_comb
```

```python
merged_genes_country = pd.merge(countries_comb, genes, on="id")
merged_genes_country
```

```python
print(f"I found on the internet which bmi is considered to be underweight, healthy, overweight, or obesitas class 1, 2 or 3. This is interesting information for the analyses, so I create a function")
def bmi_category(value):

    if value < 18.5:
        return "underweight"
    
    elif value < 25:
        return "healthy"
    
    elif value < 30:
        return "overweight"
    
    elif value < 35:
        return "obesity_class_1"
    
    elif value < 40:
        return "obesity_class_2"
    
    else:
        return "obesity_class_3"
```

```python
print(f"I will apply the function to bmi column and create a new column called bmi_category with the results\n")

merged_genes_country["bmi_category"] = merged_genes_country["bmi"].apply(bmi_category)
print(merged_genes_country)
```
